```python
target_locations = ["Black Creek", "Glenfield-Jane Heights", "York University Heights"]
filtered_df = df2[df2["NEIGHBOURHOOD_158"].isin(target_locations)]
df2= filtered_df
```

```python

import matplotlib.pyplot as plt
import seaborn as sns

data = df2

data['OCC_YEAR'].value_counts().sort_index().plot(kind='bar')
plt.title('Number of Crime Events by Year')
plt.xlabel('Year')
plt.ylabel('Number of Crime Events')
plt.show()

data['OCC_MONTH'].value_counts().sort_index().plot(kind='bar')
plt.title('Number of Crime Events by Month')
plt.xlabel('Month')
plt.ylabel('Number of Crime Events')
plt.show()


data['OCC_DOW'].value_counts().sort_index().plot(kind='bar')
plt.title('Number of Crime Events by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Number of Crime Events')
plt.show()


data['NEIGHBOURHOOD_158'].value_counts().plot(kind='barh', figsize=(10, 10))
plt.title('Number of Crime Events by Neighbourhood')
plt.xlabel('Number of Crime Events')
plt.ylabel('Neighbourhood')
plt.show()


data['OCC_TIME_RANGE'].value_counts().plot(kind='bar')
plt.title('Number of Crime Events by Time Range')
plt.xlabel('Time Range')
plt.ylabel('Number of Crime Events')
plt.show()

pivot_table = data.pivot_table(values='OBJECTID', index='OCC_DOW', columns='OCC_TIME_RANGE', aggfunc=np.size)
sns.heatmap(pivot_table)
plt.title('Heatmap of Number of Crime Events by Day of Week and Time Range')
plt.show()
```
